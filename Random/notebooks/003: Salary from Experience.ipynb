```python
sns.scatterplot(x="Years Experience", y="Salary", data=df, s=24, color="red")
plt.title("Years of Experience vs Salary", font="Hurmit Nerd Font Mono", size=24)
plt.xlabel("Years of Experience" , font="Hurmit Nerd Font Mono", size=18)
plt.ylabel("Salary", font="Hurmit Nerd Font Mono", size=18)
plt.show()
```

```python
import sklearn.linear_model as lm
 
model = lm.LinearRegression()
model.fit(df["Years Experience"].to_frame(), df["Salary"].to_frame())

x = np.array([[1, 2], [3, 4], [5, 6]])
x.shape  # (5,) â€” 1D

 
```
