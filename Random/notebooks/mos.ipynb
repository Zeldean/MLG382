```python
df = pd.read_csv(r"./../data/Monthly_Airline_Passenger_Data.csv")
df.head()
```

```python
plt.figure(figsize=(30, 10))
sns.lineplot(x="Month", y="#Passengers", data=df)
plt.title("Airline Passengers Over Time", font="Hurmit Nerd Font Mono", size=32, weight="bold")
plt.xlabel("Month", font="Hurmit Nerd Font Mono", size=22, weight="regular")
plt.ylabel("# Passengers", font="Hurmit Nerd Font Mono", size=22, weight="regular")
pollishedLabels = [f"{str(lName).split('-')[1]}({str(lName).split('-')[0]})" for lName in df["Month"][::5]]
plt.xticks(ticks=df["Month"][::5], labels=pollishedLabels, rotation=45)
plt.show()
```

```python
# Basic statistics
mean_passengers = df['#Passengers'].mean()
median_passengers = df['#Passengers'].median()
std_dev_passengers = df['#Passengers'].std()

print(f"Mean: {mean_passengers}")
print(f"Median: {median_passengers}")
print(f"Standard Deviation: {std_dev_passengers}")

# Time series statistics
min_passengers = df['#Passengers'].min()
max_passengers = df['#Passengers'].max()
range_passengers = max_passengers - min_passengers

print(f"Minimum Passengers: {min_passengers}")
print(f"Maximum Passengers: {max_passengers}")
print(f"Range of Passengers: {range_passengers}")

# Trend analysis
passenger_growth = (df['#Passengers'].iloc[-1] - df['#Passengers'].iloc[0]) / df['#Passengers'].iloc[0] * 100
print(f"Passenger Growth Over Time: {passenger_growth:.2f}%")
```
